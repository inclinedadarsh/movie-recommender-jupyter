```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
# merging the two dataframes based on their title
# could have done it with id as well, but it's name is different in both, so we'll just choose title

movies = movies.merge(credits, on='title')
```

```python
movies.head()
```

Columns we're going to take into recommendation:

- Genres
- id
- keyword
- title
- overview
- cast
- crew

```python
movies = movies[['genres', 'id', 'keywords', 'title', 'overview', 'cast', 'crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
from ast import literal_eval

def convert(obj):
    lst = []
    for i in literal_eval(obj):
        lst.append(i['name'])
    
    return lst
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    lst = []
    counter = 0
    
    for i in literal_eval(obj):
        if counter != 3:
            lst.append(i['name'])
            counter += 1
        else: break
    
    return lst
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    lst = []
    for i in literal_eval(obj):
        if i['job'] == 'Director':
            lst.append(i['name'])
            break
    return lst
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['id', 'title', 'tags']]
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.iloc[0].tags
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
ps = PorterStemmer()
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.vocabulary_
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
new_df[new_df['title'] == 'Batman Begins'].index[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
